```python
        # Equipo 6 : JUEGO DEL CAOS PARA LA GENERACIÓN DE GRÁFICOS DE LAS SECUENCIAS DE ADN

#Se importan las librerías que utilizaremos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

""" Guardamos en la serie "vértices" las coordenadas de los vértices del cuadrado. 
Cada vértice representa una base de la siguiente manera: 
A[0, 0] = Adenina, C[0, 1] = Citosina, G[1, 1] = Guanina, T[1, 0] = Timina """ 

vertices = {
    "A" : [0, 0],
    "C" : [0, 1],
    "G" : [1, 1],
    "T" : [1, 0]
}
vertices
```

```python
# Generamos un punto inicial aleatorio dentro del cuadrado junto 
punto = pd.Series([np.random.random(), np.random.random()])

punto
```

```python
""" Generamos la lista "puntos" para almacenar las coordenadas del punto inicial.
Posteriormente iremos guardando en esta lista las coordenadas de los puntos de acuerdo al orden que indique la secuencia de ADN""" 
puntos = [punto]
puntos
```

```python
# Generamos la serie "secuencia", donde se guarda la secuencia de ADN sobre la cual se generará el gráfico.
secuencia = pd.Series(np.random.choice(("A", "C", "G", "T"), 10000))
secuencia
```

```python
""" Generamos un for donde se busca el vértice de dirección, con este y el punto inicial
se calculan las coordenadas de los puntos a generar. """ 
for i in range(len(secuencia)):
    """ Elige el vértice al que se dirigirá de acuerdo a lo ordenado en "secuencia". 
    Este vértice lo transforma a coordenadas mediante lo indicado en el diccionario "vertices" """ 
    vertice = vertices[secuencia.loc[i]]
    
    # Calcula el nuevo punto a la mitad de camino entre el punto actual y el vértice elegido
    punto = (punto + vertice) / 2
    
    # Guarda el nuevo punto
    puntos.append(punto)
```

```python
# Convertimos la lista de puntos en un arreglo para graficar
puntos = np.array(puntos)

# Graficamos los puntos
plt.figure(figsize=(6, 6))
plt.scatter(puntos[:, 0], puntos[:, 1], s=0.5, color='red')
plt.title("Imágen Generada para la Secuencia 1")

# Eliminar etiquetas del eje x e y
plt.gca().set_xticks([])
plt.gca().set_yticks([])

# Guardamos los gráficos en un archivo PNG
plt.savefig("secuencia1.png")

# Mostramos la gráfica
plt.show()
```
