```python
        # Equipo 6 : JUEGO DEL CAOS PARA LA GENERACIÓN DE GRÁFICOS DE LAS SECUENCIAS DE ADN

#Se importan las librerías que utilizaremos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

""" Guardamos en la serie "vértices" las coordenadas de los vértices del cuadrado. 
Cada vértice representa una base de la siguiente manera: 
A[0, 0] = Adenina, C[0, 1] = Citosina, G[1, 1] = Guanina, T[1, 0] = Timina """ 

vertices = {
    "A" : [0, 0],
    "C" : [0, 1],
    "G" : [1, 1],
    "T" : [1, 0]
}
vertices
```

```python
# Traemos la base
df = pd.read_csv("/home/jupyter-mariel/BEDU/data/Samples2.csv")
df.head()
```

```python
# Generamos una nueva columna "secuencia_serie" donde guardemos las secuencias separadas para que cada entrada de la serie corresponda a un caracter
df["secuencia_serie"] = df['Sequence'].str.split('')
df["secuencia_serie"].head()
```

```python
# Este primer for correrá por el índice del df para ir de renglón en renglón
for j in range(1,50): 
    
    # Generamos un punto inicial aleatorio dentro del cuadrado 
    punto = pd.Series([np.random.random(), np.random.random()])
    
    """ Generamos la lista "puntos" para almacenar las coordenadas del punto inicial.
    Posteriormente iremos guardando en esta lista las coordenadas de los puntos de acuerdo al orden que indique la secuencia de ADN"""
    puntos = [punto]
    
    """Este segundo for correrá por el índice de la serie de cada secuencia para ir de letra en letra. En este se busca el vértice de dirección, con este y el punto inicial, se calculan las coordenadas de los puntos a generar. """ 
    for i in range(1,50):
        
        """ Elige el vértice al que se dirigirá de acuerdo a lo ordenado en "secuencia". 
    Este vértice lo transforma a coordenadas mediante lo indicado en el diccionario "vertices" """ 
        vertice = vertices[df["secuencia_serie"][j][i]]
        
        # Calcula el nuevo punto a la mitad de camino entre el punto actual y el vértice elegido
        punto = (punto + vertice) / 2
        
        # Guarda el nuevo punto
        puntos.append(punto)
        
        # Convertimos la lista de puntos en un arreglo para graficar
        puntos_1 = np.array(puntos)

        # Graficamos los puntos
        plt.figure(figsize=(6, 6))
        plt.scatter(puntos_1[:, 0], puntos_1[:, 1], s=0.5, color='red')

        # Eliminar etiquetas del eje x e y
        plt.gca().set_xticks([])
        plt.gca().set_yticks([])

        # Guardamos los gráficos en un archivo PNG
        name = df["ID"][j]
        plt.savefig(f"secuencia_{name}.png")
        plt.close()                    # Explicitly close the figure
```
